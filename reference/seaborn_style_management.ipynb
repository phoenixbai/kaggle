# Style Management

## Controlling figure aesthetics

```
%matplotlib inline
```

```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
np.random.seed(sum(map(ord, "aesthetics")))
```

```
# simple sine wave function
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*0.5)*(7-i)*flip)
```

```
# matplotlib defaults
sinplot()
```

```
import seaborn as sns
sinplot()
```

```
# Styling figures with axes_style() and set_style()

sns.set_style("whitegrid")
data = np.random.normal(size=(20,6)) + np.arange(6)/2
sns.boxplot(data=data)
```

```
sns.set_style("dark")
sinplot()
```

```
sns.set_style("white")
sinplot()
```

```
sns.set_style("ticks")
sinplot()
```

```
# removing spines with despine()
sinplot()
sns.despine()
```

```
# offsetting spines away from data
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
```

```
# control which spines to remove
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)
```

```
# temporarily setting figure style
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```
# overriding elements of seaborn styles
print sns.axes_style()
sns.set_style("darkgrid", {"axes.facecolor":".9"})
sinplot()
```

```
# scaling plot elements with plotting_context() and set_context()

# the four preset contexts, in order of relative size are paper, notebook, talk, poster. notebook is default.
sns.set()
sns.set_context("paper")
plt.figure(figsize=(8,6))
sinplot()
```

```
sns.set_context("talk")
plt.figure(figsize=(8,6))
sinplot()
```

```
sns.set_context("poster")
plt.figure(figsize=(8,6))
sinplot()
```

```
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2.5})
sinplot()
```
